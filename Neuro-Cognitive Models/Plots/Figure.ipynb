```python
# Plots of hierarchical posteriors for original data instead of non-paramteric bootstrap
```

```python
import numpy as np
import stanity
import sys
sys.path.append('../../')
import utils
import plotting
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from scipy import stats
```

```python
pkl = utils.load_pickle("../../save/non-boot/res_lambda_modelt.pkl")
fit = pkl['fit']
```

```python
plotting.plot_posterior(data=fit['lambdahier'], xlabel='Hierarchical lambda effect', ylabel='Density',  color=sns.husl_palette(10, h=.7), legend=['Prioritization','Non-prioritization'])
#plt.savefig('figs/lambdahier_model4.pdf')
```

```python
plotting.plot_posterior(data=fit['reshier'], xlabel='Hierarchical residual', ylabel='Density',  color=sns.husl_palette(10, h=.7), legend=['Prioritization','Non-prioritization'])
plt.savefig('figs/reshier_model4.pdf')
```

```python
plotting.plot_posterior(data=fit['alphahier'], xlabel='Boundary', ylabel='Density',  color=sns.husl_palette(10, h=.7))
plt.savefig('figs/alphahier.pdf')
```

```python
plotting.plot_posterior(data=fit['etahier'], xlabel='Drift rate variability', ylabel='Density',  color=sns.husl_palette(10, h=.7))
plt.savefig('figs/etahier.pdf')
```

```python
plotting.plot_posterior(data=fit['deltahier'], xlabel='Drift rate', ylabel='Density',  color=sns.husl_palette(10, h=.7), legend=['High coherence','Low coherence'])
plt.savefig('figs/model4_deltahier.pdf')
```

```python
fig = plt.figure(figsize=(6,4))

for i in range(15):
    sns.kdeplot(fit['n200sub'][:,i,0], alpha = .8, color='#1f77b4')
    sns.kdeplot(fit['n200sub'][:,i,1], alpha = .8, color='#ff7f0e')
    sns.kdeplot(fit['n200sub'][:,i,2], alpha = .8, color='#2ca02c')
    sns.kdeplot(fit['n200sub'][:,i,3], alpha = .8, color='#d62728')
    
sns.despine()
plt.xlabel("N200 latencies")
plt.legend(labels=['HCP','HCNP','LCP', 'LCNP'])
plt.savefig('figs/model4_indv_n200lat.pdf')
```

```python
fig = plt.figure(figsize=(6,4))
sns.kdeplot(fit['n200cond'][:,0])
sns.kdeplot(fit['n200cond'][:,1])
sns.kdeplot(fit['n200cond'][:,2])
sns.kdeplot(fit['n200cond'][:,3])
sns.despine()
plt.xlabel("N200 latencies")
plt.legend(labels=['HCP','HCNP','LCP', 'LCNP'])

for i in range(15):
    sns.kdeplot(fit['n200sub'][:,i,0], alpha=.2, color='#1f77b4')
    sns.kdeplot(fit['n200sub'][:,i,1], alpha=.2, color='#ff7f0e')
    sns.kdeplot(fit['n200sub'][:,i,2], alpha=.2, color='#2ca02c')
    sns.kdeplot(fit['n200sub'][:,i,3], alpha=.2, color='#d62728')

#plt.savefig('figs/kden200_model4.pdf')
```

```python
fig = plt.figure(figsize=(6,4))
sns.kdeplot(np.concatenate([fit['n200cond'][:,0], fit['n200cond'][:,2]]))
sns.kdeplot(np.concatenate([fit['n200cond'][:,1], fit['n200cond'][:,3]]))
for i in range(15):
    sns.kdeplot(np.concatenate([fit['n200sub'][:,i,0], fit['n200sub'][:,i,2]]), alpha=.15, color='#1f77b4')
    sns.kdeplot(np.concatenate([fit['n200sub'][:,i,1], fit['n200sub'][:,i,3]]), alpha=.15, color='#ff7f0e')
sns.despine()
plt.xlabel("Hierarchical lambda effect")
plt.legend(labels=['P','NP'])
#plt.savefig('figs/kdelambda_model4.pdf')
```

```python
fig = plt.figure(figsize=(6,4))
sns.kdeplot(fit['lambdahier'][:,0])
sns.kdeplot(fit['lambdahier'][:,1])
for i in range(15):
    sns.kdeplot(fit['lambda'][:,i,0], alpha=.15, color='#1f77b4')
    sns.kdeplot(fit['lambda'][:,i,1], alpha=.15, color='#ff7f0e')
sns.despine()
plt.xlabel("Hierarchical lambda effect")
plt.legend(labels=['Prioritization','Non-prioritization'])
#plt.savefig('figs/kdelambda_model4.pdf')
```

```python
fig = plt.figure(figsize=(6,4))
sns.kdeplot(fit['reshier'][:,0])
sns.kdeplot(fit['reshier'][:,1])
for i in range(15):
    sns.kdeplot(fit['res'][:,i,0], alpha=.15, color='#1f77b4')
    sns.kdeplot(fit['res'][:,i,1], alpha=.15, color='#ff7f0e')
sns.despine()
plt.xlabel("Hierarchical residual")
plt.legend(labels=['Prioritization','Non-prioritization'])
#plt.savefig('figs/kderes_model4.pdf')
```
